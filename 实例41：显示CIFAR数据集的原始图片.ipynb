```python
import numpy as np
from scipy.misc import imsave
```

```python
filename = 'data/cifar-10-batches-bin/test_batch.bin'

bytestream = open(filename, "rb")
buf = bytestream.read(10000 * (1+ 32 * 32 * 3))
bytestream.close()
```

```python
data = np.frombuffer(buf, dtype=np.uint8)
data = data.reshape(10000, 1 + 32*32*3)
labels_image = np.hsplit(data, [1])
labels = labels_image[0].reshape(10000)
images = labels_image[1].reshape(10000, 32, 32, 3)
```

```python
# 导出第一幅图
img = np.reshape(images[0], (3, 32, 32))
img = img.transpose(1, 2, 0)
```

```python
import pylab
print(labels[0])
pylab.imshow(img)
pylab.show()
```
